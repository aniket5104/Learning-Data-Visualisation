## Customizing the plot:

```python
batsman= pd.read_csv('C:/Users/anike/OneDrive/Desktop/Data Visualisation/Datasets/sharma-kohli.csv')
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'], linewidth=2, linestyle='--', color='red', marker='o', markersize = 6)
plt.legend(['V Kohli', 'RG Sharma'])
plt.title(' Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid(True)
```

### **Customisation Options:**
| Attribute    | Usage                                    |
| ------------ | ---------------------------------------- |
| `color`      | Line color (`'red'`, `'blue'`, etc.)     |
| `linestyle`  | Line style: `'-'`, `'--'`, `':'`, `'-.'` |
| `linewidth`  | Thickness of line                        |
| `marker`     | Marker style (`'o'`, `'s'`, `'*'`, etc.) |
| `markersize` | Size of the marker                       |
| `legend`     | Shows a legend box                       |

# **Scatter Plot:**
A scatter plot displays individual data points as dots on a 2D plane. It Helps to identify:
- Patterns
- Correlations
- Outliers
- Clusters
- Mostly used i number vs number plot.

```python
x= np.linspace(0, 50,30)
y= 3*x + np.random.randint(0, 80, 30)
plt.scatter(x,y)
```

# **Bar Chart:**

- Bivariate Analysis

```python
# Simple Bar Chart:
children = [10,20,40, 10,30]
colors= ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors, children)
```

## **Horizontal** Bar charts:

```python
plt.barh(colors, children)
```

### **Multiple** Bar Charts:

```python
df= pd.read_csv('C:/Users/anike/OneDrive/Desktop/Data Visualisation/Datasets/batsman_season_record.csv')
df
```

```python
plt.bar( df['batsman'], df['2015'], width= 0.5)
plt.bar( df['batsman'], df['2016'], width= 0.5)
plt.bar( df['batsman'], df['2017'], width= 0.5)
```

Using the above method overlaps the bar charts of each batsman. Butu still a mltiple Bar Graph can be plotted using the Below method:

```python
plt.bar(np.arange(df.shape[0])- 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0])+ 0.2, df['2017'], width=0.2)
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.xticks(rotation='vertical')
plt.show()
```

---
## **Stacked** Bar Chart:


```python
plt.bar( df['batsman'], df['2017'] ,label= '2017')
plt.bar( df['batsman'], df['2016'], bottom= df['2017'],label= '2016')
plt.bar( df['batsman'], df['2015'], bottom= df['2017']+ df['2016'],label= '2015')
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```

# **Pie Chart:**

```python
data= [23, 45, 100, 20, 40]
subjects= ['Maths', 'Science', 'English', 'Hindi', 'Social Studies']
plt.pie(data, labels=subjects)
plt.show()
```
