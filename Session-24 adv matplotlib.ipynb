# **Annotations:**

Annotations are used to label specific data points.
Syntax:<br>
`plt.annotate(text, xy=(x, y), xytext=(xtext, ytext), arrowprops=dict(...))`<br><br>
**Parameters:**
| Parameter    | Purpose                          |
| ------------ | -------------------------------- |
| `text`       | The text to display              |
| `xy`         | Point to annotate (x, y)         |
| `xytext`     | Location of the text label       |
| `arrowprops` | Dictionary to define arrow style |
| `fontsize`   | Font size of the annotation text |
| `color`      | Color of the annotation text     |

# **Subplots:**

## Understanding **figure** and **axis** object:
1. **Figure:**
- The top level container that represents entire drawing surface or canvas.
- Holds all other elements: Axes, title, legends, etc.
2. **Axes:**
- region where data is plotted
- A figure can contain **one or more** axes object.


## Creating plots using **`plt.subplot()`**

```python
fig, ax= plt.subplots(figsize=(18,10))
ax.scatter(batter['avg'],batter['strike_rate'])
ax.set_title('Player Performance')
ax.set_xlabel('Batting Average')
ax.set_ylabel('Strike Rate')
fig.show()
```

- `plt.subplots()` returns a tuple containing the figure and axes objects.

```python
fig, ax= plt.subplots(nrows= 2, ncols=1, sharex=True, figsize= (18,10))
ax[0].scatter(batter['avg'],batter['strike_rate'], color= 'orange')
ax[0].set_title('Player avg vs strike rate')
ax[0].set_xlabel('Batting Average')
ax[0].set_ylabel('Strike Rate')
ax[1].scatter(batter['avg'],batter['runs'], color= 'green')
ax[1].set_title('Player avg vs runs')
ax[1].set_xlabel('Runs')
ax[1].set_ylabel('Strike Rate')
```

### Plotting **different** types of plots in **same** figure:

```python
fig, ax= plt.subplots(nrows=2, ncols=2, figsize=(18,10))
ax[0][0].scatter(batter['avg'],batter['strike_rate'], color= 'orange')
ax[0][0].set_title('Player avg vs strike rate')
ax[0][0].set_xlabel('Batting Average')
ax[0][0].set_ylabel('Strike Rate')
ax[0][1].scatter(batter['avg'],batter['runs'], color= 'green')
ax[0][1].set_title('Player avg vs runs')
ax[0][1].set_xlabel('Runs')
ax[0][1].set_ylabel('Strike Rate')
ax[1][0].hist(batter['avg'])
ax[1][1].hist(batter['runs'], color='red')
fig.show()
```

# **3D Surface Plots:**

```python
x= np.linspace(-10,10,100)
y= np.linspace(-10,10,100)
xx, yy= np.meshgrid(x, y)
z= xx**2 + yy**2
fig= plt.figure(figsize= (12,8))
ax= plt.subplot(projection= '3d')
p= ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z= np.sin(xx) + np.cos(yy)
fig= plt.figure(figsize= (12,8))
ax= plt.subplot(projection= '3d')
p= ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```
