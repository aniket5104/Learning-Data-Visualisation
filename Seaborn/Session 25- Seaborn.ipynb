Plots Under this section:
1. scatterplot

### Scatter Plot:

```python
tips= sns.load_dataset('tips')
tips
```

Using an axes level function

```python
sns.scatterplot(data= tips, x='total_bill', y='tip', hue='sex', style='time', size='size') # ---> this is a axes level function
```

Using a figure level function.

```python
sns.relplot(data= tips, x='total_bill', y='tip', kind= 'scatter', hue='sex', style='time', size='size') # ---> this is a figure level function
# default shape of figure level function is square. altough it can be customized.
```

### **Line** Plot

```python
gap= px.data.gapminder()
temp_df=gap[gap['country']== 'India']
temp_df
```

Using Axes level function.

```python
sns.lineplot(data= temp_df, x='year', y= 'lifeExp')
```

### **Facet Plot:**
- Facet Plot will only work on figure level function i.e. relplot()
- it will not work with axis level functions.

```python
sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter', col='sex') 
```

```python
sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter',row='sex') 
```

```python
sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter', col='sex', row= 'smoker') 
```

`col_wrap=` parameter: defines the no. of graphs to be show in one column.

```python
sns.relplot(data= gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

## **Distribution Plot:**
- used for univariate analysis.
- used to find out the distribution.
- Range of the observation.
- Central tendency.
- is the data bimodal?
- are there outliers?
## Plots Under distribution plot:
1. histplot
2. kdeplot
3. rugplot

figure level funtion ---> displot<br>
axes level function ---> histplot, kdeplot, rugplot

### Plotting **Univariate** histogram:
- Using `histplot()` i.e. axis level function.

```python
sns.histplot(data= tips, x='total_bill')
```

- using `displot()` i.e. figure level function.

```python
sns.displot(data= tips, x='total_bill')
```

- the `bins=` parameter:

```python
sns.displot(data= tips, x='total_bill', bins=20, hue='sex')
```

```python
### Histogram on a categorical column:
sns.displot(data=tips, x= 'day', kind= 'hist')
# it is technically not a histogram, but a count plot.
```

- the `element=` parameter:

```python
sns.displot(data= tips, x='total_bill', bins=20, hue='sex', element= 'step')
```

```python
titanic= sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data= titanic, x='age', element= 'step' ,hue='sex', kind= 'hist')
```

- Adding **Facet** rows and columns:

```python
sns.displot(data= tips, x='tip', element= 'step' ,col='sex', kind= 'hist')
```

---
### **kdeplot:**

- Using `.kdeplot()` i.e. axis level function.

```python
sns.kdeplot(data= tips, x= 'total_bill', hue='sex', fill= True)
```

- Using `.displot()` i.e. figure level function.

```python
sns.displot(data= tips, x= 'total_bill', kind= 'kde', hue='sex', fill= True)
```

---
### **Rugplot:**
- Plot marginal distributions by drawing ticks along the x and y axis.
- This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data= tips, x= 'total_bill')
sns.rugplot(data= tips, x= 'total_bill')
```

# **Matrix** Plot:
- Heatmap
- Clustermap

## **Heatmap:**

```python
temp_df= gap.pivot(index='country', columns= 'year', values= 'lifeExp')
```

```python
sns.heatmap(temp_df)
plt.figure(figsize=(15,15))
```

- the `annot=` parameter:

```python
temp_df= gap[gap['continent']=='Europe'].pivot(index='country', columns= 'year', values= 'lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(temp_df, annot= True, linewidths=0.5, cmap='summer')
```

## **Clustermap:**
- Plot a matrix dtaset as a hierarchically-clustered heatmap.
- This function requires SciPy to be available.

```python
iris =px.data.iris()
plt.figure(figsize=(10,10))
sns.clustermap(iris.iloc[:, [0, 1, 2, 3]])
```

## How to Change the size in figure level plot:

- the `height=` and `aspect=` parameters in figure level functions:

```python
sns.displot(data= tips, x= 'total_bill', kind= 'kde', hue='sex', fill= True, height=5, aspect=1.3)
```
